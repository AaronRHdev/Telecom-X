```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('../data/processed/data_without_nulls.csv')

data
```

```python
data.info()
```

```python
# Standaring all columns
columns_to_drop = ['customerID',
                    'customer.gender',
                    'phone.PhoneService',
                    'phone.MultipleLines',
                    'internet.OnlineBackup',
                    'internet.DeviceProtection',
                    'internet.StreamingTV',
                    'internet.StreamingMovies'
                    ]

columns_to_rename =  {'Churn': 'churn',
                      'customer.SeniorCitizen': 'is_elder',
                      'customer.Partner': 'have_partner',
                      'customer.Dependents': 'have_dependents',
                      'customer.tenure': 'permanence_months',
                      'internet.InternetService': 'type_internet_service',
                      'internet.OnlineSecurity': 'have_online_security',
                      'internet.TechSupport': 'have_tech_support',
                      'account.Contract': 'type_contract',
                      'account.PaperlessBilling': 'electronic_invoice',
                      'account.PaymentMethod': 'payment_method',
                      'account.Charges.Monthly': 'charges_monthly',
                      'account.Charges.Total': 'charges_total'}

columns_to_lower = ['churn', 
                    'have_partner', 
                    'have_dependents', 
                    'type_internet_service', 
                    'have_online_security',
                    'have_tech_support', 
                    'type_contract', 
                    'electronic_invoice',
                    'payment_method' ]

columns_to_0_1 = ['churn',
                  'have_partner',
                  'have_dependents',
                  'have_online_security',
                  'have_tech_support',
                  'electronic_invoice']

# Eliminating columns not necessary
data = data.drop(columns = columns_to_drop, errors = 'ignore')

# Renaming columns
data = data.rename(columns = columns_to_rename)

# Standarizing the data
data[columns_to_lower] = data[columns_to_lower].apply(lambda x: x.str.lower())

data[columns_to_0_1] = data[columns_to_0_1].apply(lambda x: x.map({'yes': 1, 'no': 0, 'no internet service': 0}))

data['type_contract'] = data['type_contract'].map({'one year': 'anual', 'month-to-month': 'monthly'})

data
```

```python
string_cols = ['type_internet_service', 'type_contract', 'payment_method']

for col in string_cols:
    print(f"\n📊 Column(string): {col}")
    print(data[col].value_counts())
```

```python
binary_cols = ['churn', 'is_elder', 'have_partner', 'have_dependents', 'have_online_security', 'have_tech_support', 'electronic_invoice']


for col in binary_cols:
    print(f"\n📊 Column(binary): {col}")
    print(data[col].value_counts())
```

```python
data['charge_daily'] = data['charges_monthly'] / 30

data.head()
```

```python
data.to_csv('../data/processed/data_transformed.csv', index = False, sep = ',')
```
